```python
import pandas as pd 
#load data
df = pd.read_csv('C:\\Users\\HP-PC\\OneDrive\\Documents\\Housing.csv')
#show first five rows and columns
df.head()
```

```python
#discover the descriptive statistics of data
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn

#histogram of the price column

plt.figure(figsize = (10,6))
sbn.histplot(df['price'],bins = 30, kde = True)
#plt.xscale('log') 
plt.title('Distribution of Housing Prices')
plt.xlabel('Housing Prices')
plt.show()

#creation of the correlation matrix

corr_matrice = df.corr()

plt.figure(figsize = (12,8))
sbn.heatmap(corr_matrice , annot = True , cmap = 'coolwarm', linewidths = .5)
plt.title('Correlation Matrix')
plt.show
```

```python
"""we notice from the correlation matrix that we have a significant probability
   between price and area, P = 0.54 """
```

```python
# Scatter plot with two characteristics (RM et PRICE)
plt.figure(figsize = (10, 6))
sbn.scatterplot(x = 'bathrooms', y = 'price', data = df)
plt.title('Relation between the Number of Bathrooms and the Price of Housing')
plt.xlabel('number of Bathrooms')
plt.ylabel('Price of Housing')
plt.show()
```

```python
# Box plot : Number of Bathrooms
plt.figure(figsize = (8, 5))
sbn.boxplot(x = df['bathrooms'])
plt.title('Box Plot of Number of Bathrooms')
plt.show()
```

```python
#Pair plot to visualize the relation between several characteristics and housing prices
sbn.pairplot(df[['bathrooms', 'area', 'stories', 'price']])
plt.show()
```
